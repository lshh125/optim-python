```python
%load_ext autoreload
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import display, Math
from sympy.abc import x, y
from sympy import ordered, Matrix, hessian
from sympy.utilities.lambdify import lambdify

gradient = lambda f, v: Matrix([f]).jacobian(v)
```

```python
func = (1 + (x + y + 1) ** 2 * (19 - 14 * x + 3 * x ** 2 - 14 * y + 6 * x * y + 3 * y ** 2)) * (
    30 + (2 * x - 3 * y) ** 2 * (18 - 32 * x + 12 * x ** 2 + 48 * y - 36 * x * y + 27 * y ** 2))
v = list(ordered(func.free_symbols))
display(func)
display(v)

loss_fun = lambdify([v], func)
```

```python
grad = gradient(func, v)
grad_fun = lambdify([v], grad)
grad
```

```python
hess = hessian(func, v)
hess_fun = lambdify([v], hess)
hess
```

```python
%autoreload
from optim.gradient import gradient_descent

rec = gradient_descent(loss_fun, grad_fun, np.array([0.3, 0.2]), lr = 1e-4, max_iter=20)
```

```python
import plotly.graph_objects as go

xx = np.linspace(-2, 2, 200)
yy = np.linspace(-2, 2, 200)
zz = [[loss_fun([xxx, yyy]) for xxx in xx] for yyy in yy]

trace = [loss_fun(i) for i in rec]

fig = go.Figure(go.Surface(
    contours = {
        "z": {"show": True, "start": 0, "end": 10000, "size": 100}
    },
    x = xx,
    y = yy,
    z = zz,
    colorscale='Viridis'))
fig.update_layout(
        scene = {
            "xaxis": {"nticks": 20},
            "zaxis": {"nticks": 4},
            'camera_eye': {"x": 0, "y": -1, "z": 0.5},
            "aspectratio": {"x": 1, "y": 1, "z": 0.2}
        })
fig.add_scatter3d(x=rec[:, 0], y=rec[:, 1], z=trace, connectgaps=True, marker_size=2)
fig.show()
```
